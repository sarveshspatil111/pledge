```
a = [1, 2, 6, 3, 1, 4]
```

```
def not_ascending(l):
  n = 0
  for i in range(0, len(l) - 1):
    if l[i] > l[i + 1]:
      n = len(l) - (i + 1)
      break

  return n
```

```
print(not_ascending(a))
```
